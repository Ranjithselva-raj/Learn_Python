Data Visualization with Matplotlib

Matplotlib is a powerfull plotting library for python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of matplotlib, including creating various types of plots and custmizing them. 

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,7,6,8,10]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line graph")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,7,6,8,10]

#create customize line plot
plt.plot(x,y, color = "red", linewidth = 2, linestyle = ":",marker = "o")
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line graph")
plt.show()
```

```python
## Create multiple plots
## Sample data

x = [1,2,3,4,5]
y = [2,7,6,8,10]
y1 = [3,9,5,7,11]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y, color ='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y,x, color ='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y1, color ='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y1,x, color ='purple')
plt.title("Plot 4")
```

```python
##bar plot
categories = ['Frogs', 'Hogs', 'Dogs', 'Logs']
values = [15, 30, 45, 10]
##create a bar plot
plt.bar(categories, values, color ='maroon')
plt.grid(visible=False)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar plot")

plt.show()
```

Histograms

Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
##Histograms
##sample data

data = [1,3,2,4,4,5,6,6,7,7,8,8,9,9,1,2,3,4,5,3,2,2,3,4,3,2,2,5,10]

#create a histogram
plt.hist(data, bins = 10, color = 'green',edgecolor = 'black')
plt.show() 
```

```python
##scatter plot

x = [1,2,3,4,5]
y = [2,7,6,8,10]

#create a scatter plot
plt.scatter(x,y, color = 'green', marker = '*',s = 100)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
##pie chart

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
values = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(values, explode=explode, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=50)
plt.axis('equal')
plt.show()
```

```python
## Sales data visualization
import pandas as pd
sales_data = pd.read_csv("sales_data.csv")
sales_data.head(5)
```

```python
total_sales_by_product = sales_data.groupby("Product Category")['Total Revenue'].sum()

print(total_sales_by_product)

total_sales_by_product.plot(kind = 'bar')
```

```python
total_sales_by_region = sales_data.groupby("Region")['Total Revenue'].sum()

total_sales_by_region.plot(kind = 'bar')
```

```python
##plot sales trend over time
sales_data_by_date = sales_data.groupby("Date")['Total Revenue'].sum().reset_index() #reset_index() is used to convert the index into a column

plt.plot(sales_data_by_date['Date'], sales_data_by_date['Total Revenue'])
```
