Data Visualization with Seaborn

Seaborn is a python visualization library based on matplotlib that provides a high level interface for drawing attractive and informative statistical grapics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#basic plotting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total_bill vs tip")
##show the plot
plt.show()
```

```python
#line plot

sns.lineplot(x = 'size', y='total_bill',data=tips)
plt.title("Line plot of total_bill vs size")
plt.show()
```

```python
#categorical variable plot
#bar plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total_bill by day")
plt.show()
```

```python
#box plot

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot of total_bill by day")
plt.show()
```

```python
##violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin plot of total_bill by day")
plt.show()
```

```python
##histogram

sns.histplot(x='total_bill',bins=10,kde=True,data=tips)
plt.title("Histogram of total_bill")
plt.show()
```

```python
## kde plot
sns.kdeplot(x='total_bill',fill=True,data=tips)
plt.title("KDE plot of total_bill")
plt.show()
```

```python
##pair plot
sns.pairplot(tips)
plt.show()
##pair plot
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
##heatmap
corr=tips[['total_bill','tip','size']].corr()
print(corr)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
#customization of plot

sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
#plot total sales by product

plt.figure(figsize=(10,5))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum,color='orange')
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#plot total sales by region

plt.figure(figsize=(10,5))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum,color='orange')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```
