```python
products_list = [{"product":"shoes", "price": 1000,"description": "sneakers"}, 
                 {"product":"t-shirt", "price": 500,"description": "casual"},
                   {"product":"jeans", "price": 2000,"description": "casual"},
                   {"product":"sunglasses", "price": 500,"description": "fashion"},
                   {"product":"Monitor", "price": 1000,"description": "LED"}]


cart =[]
while True:
    choice = input("Do you want to add an item to the cart? (y/n): ")
    if choice.lower() == "y":
        print("Here is the list of products")

        for index, product in enumerate(products_list):
            print(f"{index}: {product['product']} : £{product['price']}: {product['description']}")

        user_purchase_id = int(input("Enter id of item you want to purchase: "))

        #check if product already present in the cart and create increment of quantity

        if products_list[user_purchase_id] in cart:
            products_list[user_purchase_id]["quantity"] += 1
        else:
            products_list[user_purchase_id]["quantity"] = 1
            cart.append(products_list[user_purchase_id])
        
        total = 0
        print(f"Current cart contents are")

        #displaying the cart content which user can understand
        for product in cart:
            print(f"{product['product']} : £{product['price']} :{product['quantity']}")
            total = total + product['price'] * product['quantity']

        print(f"Cart Total: £{total}")

    else:
        break

#print(f"Thank you. Your final cart is {cart} with total items {len(cart)}")

def display_final_cart():
    for product in cart:
            print(f"{product['product']} : £{product['price']} :{product['quantity']}")
            print(f"Cart Total: £{total}")

display_final_cart()
```
